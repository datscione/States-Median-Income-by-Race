```python
file = pd.read_csv('ACSST5Y2019.S1903-2021-06-03T012656.csv')
file.head(20)
```

```python
df = file.iloc[[4,6,10,11]]
df.head()
```

```python
df.columns = df.columns.str.replace('!!', ' ')
df.columns
```

```python
df.head()
```

```python
margin_df = df[df.columns[::2]]
margin_df
```

```python
margin_df.columns
```

```python
med_inc = df.copy()
med_inc = med_inc.drop(labels=margin_df, axis=1)
med_inc.head()
```

```python
margin_df['Label'].head()
```

```python
med_inc.insert(0, 'Ethnicity', margin_df['Label'])
med_inc.head()
```

```python
remove_comma = med_inc[med_inc.columns[1:]].replace(',','', regex=True)
remove_comma.head()
```

```python
df_num = remove_comma.select_dtypes(object).columns
remove_comma[df_num] = remove_comma[df_num].apply(pd.to_numeric,errors='coerce')
```

```python
type(remove_comma['Alabama Median income (dollars) Estimate'].iloc[0])
```

```python
med_inc[med_inc.columns[1:]] = remove_comma
med_inc.info()
```

```python
med_inc.describe()
```

```python
med_inc = med_inc.set_index('Ethnicity')
med_inc.head()
```

```python
pd.options.display.max_columns = None
states_rep_columns = ['California Median income (dollars) Estimate', 'Florida Median income (dollars) Estimate',
                  'Illinois Median income (dollars) Estimate', 'New York Median income (dollars) Estimate', 
                  'Texas Median income (dollars) Estimate', 'Washington Median income (dollars) Estimate', 
                  'District of Columbia Median income (dollars) Estimate', 'Maryland Median income (dollars) Estimate', 
                  'Massachusetts Median income (dollars) Estimate', 'North Carolina Median income (dollars) Estimate', 
                  'Ohio Median income (dollars) Estimate','Colorado Median income (dollars) Estimate', 
                  'Michigan Median income (dollars) Estimate','Indiana Median income (dollars) Estimate', 
                  'Nevada Median income (dollars) Estimate','Kentucky Median income (dollars) Estimate', 
                  'Tennessee Median income (dollars) Estimate','Oklahoma Median income (dollars) Estimate', 
                  'Pennsylvania Median income (dollars) Estimate','Arizona Median income (dollars) Estimate', 
                  'Oregon Median income (dollars) Estimate','Wisconsin Median income (dollars) Estimate', 
                  'New Mexico Median income (dollars) Estimate', 'Georgia Median income (dollars) Estimate', 
                  'Missouri Median income (dollars) Estimate', 'Nebraska Median income (dollars) Estimate']
rep_states = med_inc[states_rep_columns]
rep_states
```

```python
rep_states.idxmin()
```

```python
rep_states.idxmax()
```

```python
from collections import Counter
```

```python
race_min_labels = Counter(rep_states.idxmin()).keys()
```

```python
x_min_vals = Counter(rep_states.idxmin()).values()
```

```python
fig, ax = plt.subplots()
ax.pie(x_min_vals, labels=race_min_labels, autopct='%1.1f%%')
ax.axis('equal')
plt.title('Lowest Median Income by Race of Most Populated U.S. States and D.C.')
```

```python
race_max_labels = Counter(rep_states.idxmax()).keys()
```

```python
x_max_vals = Counter(rep_states.idxmax()).values()
```

```python
fig, ax = plt.subplots()
ax.pie(x_max_vals, labels=race_max_labels, autopct='%1.1f%%')
ax.axis('equal')
plt.title('Highest Median Income by Race of Most Populated U.S. States and D.C.')
```

```python
min_med_inc_rep= rep_states.min(axis=1)
min_med_inc_rep
```

```python
max_med_inc_rep = rep_states.max(axis=1)
max_med_inc_rep
```

```python
# fig1 = plt.figure()
# plt.plot(range(10), range(10), "o")
# fig1.savefig("fig1.png", bbox_inces="tight", dpi=550)
```

```python
min_inc_rep_states = rep_states.min(axis=0)
min_inc_rep_states
```

```python
max_inc_rep_states = rep_states.max(axis=0)
max_inc_rep_states
```

```python
inc_disparity_rep = (max_inc_rep_states - min_inc_rep_states).sort_values(ascending=True)
inc_disparity_rep
```

```python
plt.barh(inc_disparity_rep.index, inc_disparity_rep)
plt.title('Median Income Disparity')
```

D.C. had the highest income disparity (approximately $100,000), followed by Illinois, Massachusetts, and Michigan. Their income disparities were nearly tied and were half that of D.C. The lowest income disparity was in Oklahoma.

```python
x_indices = np.arange(len(min_med_inc_rep.index))
width = 0.3

plt.bar(x_indices, min_med_inc_rep, width=width)
plt.bar(x_indices+width, max_med_inc_rep, width=width)

plt.xticks(ticks=x_indices, labels=min_med_inc_rep.index, rotation=40, horizontalalignment='right')
plt.legend()

plt.title('Lowest and Highest Median Incomes by Race of Most Populated U.S. States and D.C.')
```

Whites have the highest median income disparity and the highest median income. Their lowest median income was 52387 and the highest was 141650. Asians had the 'highest' low median income and blacks had the lowest overall income. 

Below is the income analysis excluding D.C. since its income disparity is twice that of the next values.

```python
nodc_df = rep_states.drop(labels='District of Columbia Median income (dollars) Estimate', axis=1)
nodc_df
```

```python
nodc_min = nodc_df.min(axis=1)
nodc_min
```

```python
nodc_max = nodc_df.max(axis=1)
nodc_max
```

```python
x_indices = np.arange(len(nodc_min.index))
width = 0.3

plt.bar(x_indices, nodc_min, width=width)
plt.bar(x_indices+width, nodc_max, width=width)

plt.xticks(ticks=x_indices, labels=nodc_min.index, rotation=40, horizontalalignment='right')
plt.legend()

plt.title('Lowest and Highest Median Incomes by Race of Most Populated U.S. States')
```

Asians have the highest income when D.C. is excluded. This means there are whites in the D.C. area whose income is much more than those in other parts of the U.S. 

```python
nodc_df.idxmin()
```

```python
nodc_df.idxmax()
```

```python
nodc_race_min_labels = Counter(nodc_df.idxmin()).keys()
nodc_x_min_vals = Counter(nodc_df.idxmin()).values()
fig, ax = plt.subplots()
ax.pie(nodc_x_min_vals, labels=nodc_race_min_labels, autopct='%1.1f%%')
ax.axis('equal')
plt.title('Lowest Median Income by Race of Most Populated U.S. States')
```

```python
nodc_race_max_labels = Counter(nodc_df.idxmax()).keys()
nodc_x_max_vals = Counter(nodc_df.idxmax()).values()
fig, ax = plt.subplots()
ax.pie(nodc_x_max_vals, labels=nodc_race_max_labels, autopct='%1.1f%%')
ax.axis('equal')
plt.title('Highest Median Income by Race of Most Populated U.S. States')
```

When D.C. is excluded, Asians account for 92% of the highest median income and whites 8%.

This study shows that there is great income disparity in the most populated parts of the U.S. In the nation's capitol especially, there is the greatest income disparity where whites account for most of the high income. Nationally, Asians have the highest income in the top 40 most populated areas, with the exception of New York in which whites had the highest income. Blacks have the lowest income in these regions, besides Massachusetts where hispanics had the lowest. The analysis below considers all states and Puerto Rico.
